```python
import pandas as pd
import seaborn as sns
import math
# Charger le dataset Titanic
df = sns.load_dataset ("titanic")
# On retire les lignes ou l‚Äôage est inconnu pour simplifier les calculs
# C‚Äôest notre Univers S.
df = df.dropna (subset = ["age"])
```

### Exercise 1

```python
# 1. D√©finition des √©v√©nements :
# A = "le passager a surv√©cu" (survived == 1)
# B = "le passager est en premi√®re classe" (pclass == 1)

# 3. Trouver |B| = nombre total de passagers en premi√®re classe
nb_first_class = df[df.pclass == 1].shape[0]
print(f"3. Nombre total de passagers en premi√®re classe (|B|) = {nb_first_class}")

# 4. Trouver |A ‚à© B| = nombre de survivants en premi√®re classe
nb_survived_first_class = df[(df.pclass == 1) & (df.survived == 1)].shape[0]
print(f"4. Nombre de survivants en premi√®re classe (|A ‚à© B|) = {nb_survived_first_class}")

# 5. Calcul de P(A|B) = |A ‚à© B| / |B|
p_survived_first_class = nb_survived_first_class / nb_first_class
print(f"5. Probabilit√© qu‚Äôun passager ait surv√©cu sachant qu‚Äôil est en 1√®re classe (P(A|B)) = {p_survived_first_class:.4f}")
```

### Exercice 2

```python
# 1. Trouver |S| (nombre total de passagers) et |A1| (nombre total de survivants)
total_passengers = df.shape[0]
nb_survivors = df[df.survived == 1].shape[0]
print(f"1. |S| = nombre total de passagers = {total_passengers}")
print(f"   |A1| = nombre total de survivants = {nb_survivors}")

# 2. Calculer P(A1) = |A1| / |S|
p_A1 = nb_survivors / total_passengers
print(f"2. P(A1) = probabilit√© que le premier passager soit un survivant = {p_A1:.4f}")

# 3. Calculer P(A2|A1) = (nb_survivors - 1) / (total_passengers - 1)
p_A2_given_A1 = (nb_survivors - 1) / (total_passengers - 1)
print(f"3. P(A2|A1) = probabilit√© que le second passager soit un survivant sachant que le premier l‚Äô√©tait = {p_A2_given_A1:.4f}")

# 4. Calculer le produit P(A1 ‚à© A2) = P(A1) √ó P(A2|A1)
p_two_successive_survivors = p_A1 * p_A2_given_A1
print(f"4. P(A1 ‚à© A2) = probabilit√© que les deux passagers tir√©s soient des survivants = {p_two_successive_survivors:.6f}")
```

### Exercice 3

```python
# ‚öôÔ∏è Donn√©es de base
total_passengers = df.shape[0]
nb_survivors = df[df.survived == 1].shape[0]

# Comptage par classe
nb_first_class = df[df.pclass == 1].shape[0]
nb_survived_first_class = df[(df.pclass == 1) & (df.survived == 1)].shape[0]

nb_second_class = df[df.pclass == 2].shape[0]
nb_survived_second_class = df[(df.pclass == 2) & (df.survived == 1)].shape[0]

nb_third_class = df[df.pclass == 3].shape[0]
nb_survived_third_class = df[(df.pclass == 3) & (df.survived == 1)].shape[0]  # üîß correction ici (√©tait 1 au lieu de 3)

# 1Ô∏è‚É£ Calcul des P(Ai)
p_A1 = nb_first_class / total_passengers
p_A2 = nb_second_class / total_passengers
p_A3 = nb_third_class / total_passengers
print(f"1. P(A1) = probabilit√© d‚Äô√™tre en 1√®re classe = {p_A1:.4f}")
print(f"   P(A2) = probabilit√© d‚Äô√™tre en 2e classe   = {p_A2:.4f}")
print(f"   P(A3) = probabilit√© d‚Äô√™tre en 3e classe   = {p_A3:.4f}")

# 2Ô∏è‚É£ Calcul des P(B|Ai)
p_B_given_A1 = nb_survived_first_class / nb_first_class
p_B_given_A2 = nb_survived_second_class / nb_second_class
p_B_given_A3 = nb_survived_third_class / nb_third_class
print(f"2. P(B|A1) = probabilit√© de survie en 1√®re classe = {p_B_given_A1:.4f}")
print(f"   P(B|A2) = probabilit√© de survie en 2e classe   = {p_B_given_A2:.4f}")
print(f"   P(B|A3) = probabilit√© de survie en 3e classe   = {p_B_given_A3:.4f}")

# 3Ô∏è‚É£ Formule des probabilit√©s totales
p_B_total = (
    p_B_given_A1 * p_A1 +
    p_B_given_A2 * p_A2 +
    p_B_given_A3 * p_A3
)
print(f"3. P(B) selon la formule des probabilit√©s totales = {p_B_total:.4f}")

# 4Ô∏è‚É£ V√©rification : comparaison avec le calcul direct
p_B_direct = nb_survivors / total_passengers
print(f"4. P(B) calcul direct = {p_B_direct:.4f}")

# ‚úÖ V√©rification de coh√©rence
print(f"‚úÖ V√©rification : diff√©rence entre les deux m√©thodes = {abs(p_B_total - p_B_direct):.6f}")
```
