```python
import pandas as pd
import seaborn as sns
import math
# Charger le dataset Titanic
df = sns.load_dataset ("titanic")
# On retire les lignes ou l’age est inconnu pour simplifier les calculs
# C’est notre Univers S.
df = df.dropna (subset = ["age"])
```

### Exercice 1:

```python
#Question 1
size_df = df.shape[0]
print(f"Taille du DataFrame (|S|): {size_df}")

#Question 2
event_a = df[(df.pclass == 3) & (df.age > 40) & (df.sex == "male")].shape[0]
print(f"Cardinal (taille) de l'ensemble correspondant à l'évenement |A|: {event_a}")

#Question 2
event_a_probability = round(event_a/size_df,2)*100
print(f"Probabilité l'évenement |A|: {event_a_probability}%")
```

### Exercice 2:

```python
nb_people_southampton = df[df.embark_town == "Southampton"].shape[0]
print(f"Nombre de passagers qui ont embarqué à Southampton (n): {nb_people_southampton}")

k = 5
print(f"Taille des comités que l'on veut former: {k}")

nb_possible_combs = math.comb(nb_people_southampton, k)
print(f"Nombre de commités possibles: {nb_possible_combs}")
```

### Exercice 3

```python
nb_childrens = df[df.age <= 12].shape[0]
print(f"Nombre d'enfants qui ont embarqué: {nb_childrens}")

k = 4
print(f"Taille des comités que l'on veut former: {k}")

nb_possible_orders = math.perm(nb_childrens, k)
print(f"Nombre de d'ordres possibles: {nb_possible_orders}")
```

### Exercice 4

```python
total_passengers = df.shape[0]
print(f"1. |S| (nombre total de passagers) = {total_passengers}")

passenger_survived = df[df.survived == 1].shape[0]
print(f"2. |A| (nombre de survivants) = {passenger_survived}")

first_class_passenger = df[df.pclass == 1].shape[0]
print(f"3. |B| (nombre de passagers en 1ère classe) = {first_class_passenger}")

survived_and_first_passenger = df[(df.survived == 1) & (df.pclass == 1)].shape[0]
print(f"4. |A ∩ B| (nombre de survivants en 1ère classe) = {survived_and_first_passenger}")

p_a = passenger_survived / total_passengers
p_b = first_class_passenger / total_passengers
p_a_inter_b = survived_and_first_passenger / total_passengers
p_a_union_b = p_a + p_b - p_a_inter_b

print(f"5. P(A ∪ B) = P(A) + P(B) − P(A ∩ B) = {p_a_union_b:.4f}")
```

### Exercise 5

```python
s = df.shape[0]
print(f"1. |S| (nombre total de passagers) = {s}")

m= df[df.survived == 1].shape[0]
print(f"2. m (nombre de survivants) = {m}")

p = s-m
print(f"3. p (nombre de morts) = {p}")

combs = math.comb(s,10)
print(f"4. Combinaisons possibles de 10 passagers= {combs}")

specific_combs = math.comb(m,4)*math.comb(p,6)
print(f"5. Combinaisons possibles de 4 survivants et 6 morts = {specific_combs}")

p_specific_combs = specific_combs/combs
print(f"6. Porbabilité de choisir 4 survivants et 6 morts = {specific_combs/combs*100:.2f}%")
```

```python
nb_passengers = df.shape[0]
nb_alone_passengers = df[df.alone == True].shape[0]

total_five_combs = math.comb(s,5)
five_combs_without_alone = math.comb(nb_passengers - nb_alone_passengers,5)

print(f"1. Probabilité de de sélectionner 5 passagers qui ne voyagent pas seuls = {five_combs_without_alone/total_five_combs*100:.2f}%")
print(f"1. Probabilité de de sélectionner 5 passagers dont au moins 1 qui voyage seul = {(1 - five_combs_without_alone/total_five_combs)*100:.2f}%")
```

### Exercice 7

```python
# 1. Calcul du tarif moyen
mean_fare = df.fare.mean()
print(f"1. Tarif moyen de tous les passagers (mean_fare) = {mean_fare:.2f}")

# 2. Taille totale de l'univers |S|
nb_passengers = df.shape[0]
print(f"2. Taille totale de l'univers (|S|) = {nb_passengers}")

# 3. Nombre de passagers dont le fare est supérieur au tarif moyen |A|
nb_passengers_superior_fare = df[df.fare > mean_fare].shape[0]
print(f"3. Nombre de passagers avec un tarif supérieur à la moyenne (|A|) = {nb_passengers_superior_fare}")

# 4. Calcul de la probabilité P(A)
p_superior_fare = nb_passengers_superior_fare / nb_passengers
print(f"4. Probabilité qu’un passager ait payé un tarif supérieur à la moyenne (P(A)) = {p_superior_fare:.4f}")
```
