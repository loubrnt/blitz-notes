```python
import pandas as pd
import seaborn as sns
import math

df = sns.load_dataset("taxis")
df.dropna(subset="passengers", inplace=True)
df = df[df.passengers > 0]
```

### Exercice 1

```python
p_X1 = df[df.passengers==1].shape[0]/df.shape[0]
p_X2 = df[df.passengers==2].shape[0]/df.shape[0]
p_X3 = df[df.passengers==3].shape[0]/df.shape[0]
p_X4 = df[df.passengers==4].shape[0]/df.shape[0]
p_X5 = df[df.passengers==5].shape[0]/df.shape[0]
p_X6 = df[df.passengers==6].shape[0]/df.shape[0]

print(sum([p_X1,p_X2,p_X3,p_X4,p_X5,p_X6]))
```

### Exercice 2

```python
P_X_inf_equal_2 = p_X1 + p_X2
print(f"{P_X_inf_equal_2*100:.2f}%")
P_X_sup_3 = 1 - p_X1 - p_X2 - p_X3
print(f"{P_X_sup_3*100:.2f}%")
```
